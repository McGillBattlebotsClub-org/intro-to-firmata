# Example of using the keep_alive method of Pymata4
The arduino pin13 LED will toggle on/off until the connection to the host PC is down. If that happens, the arduino resets itself, therefore turning off the pin13 LED

```python
import sys
import time

from pymata4 import pymata4


# Setup a pin for digital output
LED_PIN = 13

def setup(my_board):
    """
    This function will toggle a digital pin.
    :param my_board: an PymataExpress instance
    """
    
    # The arduino will reset itself after 5 seconds of being disconnected to the host PC
    my_board.keep_alive(period=5) 
    
    # set the pin mode
    my_board.set_pin_mode_digital_output(LED_PIN)
    
    # toggle the pin forever
    while True:
        board.digital_write(LED_PIN, 0)
        time.sleep(1)
        board.digital_write(LED_PIN, 1)
        time.sleep(1)
        
    my_board.shutdown()

    
board = pymata4.Pymata4("COM6")
try:
    setup(board)
except KeyboardInterrupt:
    board.shutdown()
    sys.exit(0)
```
